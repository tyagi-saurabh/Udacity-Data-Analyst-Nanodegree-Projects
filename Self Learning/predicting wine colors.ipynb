```python
# building an ANN-I
# Classification template

# Importing the libraries

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Importing the dataset
dataset = pd.read_csv('winequality_edited.csv')
```

```python
# getting the info on dataset
dataset.info() # the last column is missing one value, we will drop it
```

```python
# dropping the na data
dataset.dropna(inplace = True)
```

```python
# checking for changes
dataset.info()
```

```python
# dropping columns
dataset.drop('acidity levels', axis = 1, inplace = True)
```

```python
# checking for changes
dataset.info()
```

```python
# changing colors to dummy data
dataset[['red','white']] = pd.get_dummies(dataset['color'])
dataset.drop(['color', 'white'], axis = 1, inplace = True)
dataset.info()
```

```python
# selecting the independent and dependent data
X = dataset.iloc[:, 0:12].values
Y = dataset.iloc[:, 12].values
Y
```

```python
# Splitting the dataset into the Training set and Test set
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 0)
```

```python
# Feature Scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
import os
os.environ['KERAS_BACKEND'] = 'tensorflow'

import keras

# building the netwrok
from keras.models import Sequential # initialises the model
from keras.layers import Dense
```

```python
# initilaising the ann
classifier = Sequential() # we named our neural network classifier since it is a classification problem
```

```python
# add the input layer and the hidden layer
classifier.add(Dense(7, activation = 'relu', input_dim = 12, kernel_initializer = 'uniform'))
```

```python
# adding the second hidden layer
classifier.add(Dense(7, activation = 'relu',  kernel_initializer = 'uniform'))
```

```python
# this network consists of 2 hidden layers
# here we are adding the final output layer
classifier.add(Dense(1, activation = 'sigmoid',  kernel_initializer = 'uniform'))
```

```python
# compile our ann
# optimizer finds the best weights
# loss is the loss function which optimises weights by minimizing loss
classifier.compile(optimizer = 'rmsprop', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

```python
# fit the ann to training set
classifier.fit(X_train, Y_train, batch_size = 64, epochs = 100 )
```

```python
# testing model performance
test_loss, test_accuracy = classifier.evaluate(X_test, Y_test)
print('test loss: ', test_loss)
print('test accuracy: ', test_accuracy) # alternate way of verifying the accuracy of the model
```

```python
# making the predictions and evaluating the model

# predicting the test set results
Y_pred = classifier.predict(X_test)
```

```python
# viewing the results
Y_pred
```

```python
Y_pred[-1] # the probability that the last wine is a red wine is 0.0002 i.e it's a white wine
```

```python
# making the confusion matrix

# changing the probabilities to boolean values
Y_pred = Y_pred>0.90

# importing the confusion matrix
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(Y_test, Y_pred)
```

```python
# Evaluating the ann
from keras.wrappers.scikit_learn import KerasClassifier # wrapper which wraps around scikitlearn library
from sklearn.model_selection import cross_val_score

def build_classifier():
    classifier = Sequential()
    classifier.add(Dense(7, activation = 'relu', input_dim = 12, kernel_initializer = 'uniform'))
    classifier.add(Dense(7, activation = 'relu',  kernel_initializer = 'uniform'))
    classifier.add(Dense(1, activation = 'sigmoid',  kernel_initializer = 'uniform'))
    classifier.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
    return classifier

classifier = KerasClassifier(build_fn = build_classifier, batch_size = 64, epochs = 50)
accuracies = cross_val_score(estimator = classifier, X = X_train, y = Y_train, cv = 10, n_jobs = -1)
```

```python
# taking the mean of all accuracies
accuracies.mean()
```

```python
# testing for variance
accuracies.var() # a low variance of .0000005
    
```

```python
# tuning the ann to achieve greater accuracies
from keras.wrappers.scikit_learn import KerasClassifier # wrapper which wraps around scikitlearn library
from sklearn.model_selection import GridSearchCV # this class actually helps in tuning the hyper parameters

def build_classifier(optimizer):
    classifier = Sequential()
    classifier.add(Dense(7, activation = 'relu', input_dim = 12, kernel_initializer = 'uniform'))
    classifier.add(Dense(7, activation = 'relu',  kernel_initializer = 'uniform'))
    classifier.add(Dense(1, activation = 'sigmoid',  kernel_initializer = 'uniform'))
    classifier.compile(optimizer = optimizer, loss = 'binary_crossentropy', metrics = ['accuracy'])
    return classifier

classifier = KerasClassifier(build_fn = build_classifier)
parameters = {'batch_size':[64,128], 'epochs':[50,100],'optimizer':['adam','rmsprop']}
grid_search = GridSearchCV(estimator = classifier, param_grid = parameters, scoring = 'accuracy', cv = 10)
grid_search = grid_search.fit(X_train, Y_train)
best_parameters = grid_search.best_params_
best_accuracy = grid_search.best_score_
```

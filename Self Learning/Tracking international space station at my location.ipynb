```python
import math
import time
from datetime import datetime
import ephem
degrees_per_radian = 180.0 / math.pi
home = ephem.Observer()
home.lon = '77.754' # E
home.lat = '28.9424' # N
home.elevation = 225
```

```python
# A two-line element set (TLE) is a data format encoding a list of orbital elements of an Earth-orbiting object for a given point in time, the epoch. Using suitable prediction formula, the state (position and velocity) at any point in the past or future can be estimated to some accuracy
iss = ephem.readtle('ISS',
    '1 25544U 98067A   18353.02973733  .00016717  00000-0  10270-3 0  9023' ,
'2 25544  51.6372 177.5636 0004861 158.6104 201.5252 15.54067346 27259'
)
```

```python
while True:
    home.date = datetime.utcnow()
    iss.compute(home)
    print('iss: altitude %4.1f deg, azimuth %5.1f deg' % (iss.alt * degrees_per_radian, iss.az * degrees_per_radian))
    time.sleep(1.0)
```

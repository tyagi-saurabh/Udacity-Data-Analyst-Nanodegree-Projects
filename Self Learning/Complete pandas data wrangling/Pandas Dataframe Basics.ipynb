# Subsetting rows

Subset Rows by index label: loc

```python
# let's look at part of our gapminder data
df.head(5)
```

on the left side of the printed dataframe, we see what appear to be row numbers.This column less row of values is the index of the dataframe

we can use the loc attrinute on the dataframe to subset rows based on index label

```python
# get the first row
# Python counts from 0
df.loc[0]
```

Subsetting multiple rows

```python
# select the first ,100th , and 150th rows
# note the double square brackets similar to the concept used to subset multiple columns
df.loc[[0 ,99 ,149]]
```

# Mixing it up

The loc and iloc attributes can be used to ontain sibsets of columns,rows or both. The general syntax is df.loc[[rows] ,[columns]]
or df.iloc[[rows] ,[columns]]

```python
# subset all rows and column index 5
df.iloc[: ,[5]]
```

# frequency counts

Another common data-related task is to calculate frequencies. We can use nuniqe and value_counts methods,respectively to get counts of unique values
and frequency counts on pandas series

```python
# use the unique to calculate number of unique values by continent for co2emissions
df.groupby('country')['co2emissions'].nunique()
```

```python
# try applying value_counts  to the same
```

```python
df.groupby('country')['co2emissions'].value_counts()
```

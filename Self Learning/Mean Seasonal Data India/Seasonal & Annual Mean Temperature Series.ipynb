# Proof of rising temperatures(Global warming)
The indian subcontinent is experiencing a rise in temperatures inline with the global rising tempratues. Since 1901, the mean annual mean tempratures have risen by 1 degree celsius. We are also experiencing a rise in anuual winter temperatues which means warmer winters and a fall in crop output. If we are not able to curb this trend, the day is not far when there will be a costal tsunamis(due to rise in sea levels) ,food shortage due to fall in the food crop output and a failed future for our coming generations.

```python
# importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('darkgrid')
%matplotlib inline
```

```python
# loading dataset
df = pd.read_csv('Mean_Seasonal_IMD_2017.csv')
df.head()
```

```python
# getting a brief overview of the data
df.describe()
```

```python
# proof of rising temperatures
df['Moving_average_annual'] = df['ANNUAL'].rolling(30).mean().dropna()
```

```python
# visual proof
plt.plot(df.YEAR ,df['Moving_average_annual'] ,label = 'Annual rolling mean temprature\n from 1901 to 2017')
plt.legend()
plt.xlabel('Years')
plt.ylabel('Tempratures')
plt.title('India"s rising temperature since 1901')
plt.show()
```

```python
# proof of ever warming indian summers
df['Moving_average_mar_may'] = df['MAR-MAY'].rolling(30).mean().dropna()
df['Moving_average_june_sep'] = df['JUN-SEP'].rolling(30).mean().dropna()
```

```python
# visual proofs
plt.figure
plt.plot(df.YEAR ,df['Moving_average_mar_may'] ,label = 'Annual rolling mean mar_may temprature\n from 1901 to 2017')
plt.plot(df.YEAR ,df['Moving_average_june_sep'] ,label = 'Annual rolling mean june_sep temprature\n from 1901 to 2017')
plt.legend()
plt.xlabel('Years')
plt.ylabel('Tempratures')
plt.title('Indian summers since 1901')
plt.show()
```

```python
# proof of even warmer indian winters
df['Moving_average_jan_feb'] = df['JAN-FEB'].rolling(30).mean().dropna()
df['Moving_average_oct_dec'] = df['OCT-DEC'].rolling(30).mean().dropna()
```

```python
# visual proofs
plt.plot(df.YEAR ,df['Moving_average_jan_feb'] ,label = 'Annual rolling mean jan_feb temprature\n from 1901 to 2017')
plt.plot(df.YEAR ,df['Moving_average_oct_dec'] ,label = 'Annual rolling mean oct_dec temprature\n from 1901 to 2017')
plt.legend()
plt.xlabel('Years')
plt.ylabel('Tempratures')
plt.title('Indian winters since 1901')
plt.show()
```

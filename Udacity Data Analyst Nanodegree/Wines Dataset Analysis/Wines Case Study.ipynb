# Appending Data
You can combine the **red** and **white** datasets to make your analysis more efficient. Use NumPy to create a new column that preserves color information, and then use pandas to combine the dataframes.



```python
color_red = np.repeat('red' , df_red_wine.shape[0]) # repeat function has got two parameters viz 'variable to be repeated' , 
                                                    # 'number of repeatations'
```

```python
color_white = np.repeat('white' , df_white_wine.shape[0])
```

# Drawing Conclusions Using Groupby
In the notebook below, you're going to investigate two questions about this data using pandas' groupby function. Here are tips for answering each question:

Q1: Is a certain type of wine (red or white) associated with higher quality?
For this question, compare the average quality of red wine with the average quality of white wine with groupby. To do this group by color and then find the mean quality of each group.
**white wine is of better quality**

```python
df_wine.groupby('color').mean() # seems like white wine is of somewhat better quality
```

**What level of acidity receives the highest average rating?**

__You can create a categorical variable from a quantitative variable by creating your own categories. pandas' cut function let's you "cut" data in groups. Using this, create a new column called acidity_levels with these categories:__

Acidity Levels:

High: Lowest 25% of pH values

Moderately High: 25% - 50% of pH values

Medium: 50% - 75% of pH values

Low: 75% - max pH value

```python
# View the min, 25%, 50%, 75%, max pH values with Pandas describe
df_wine.describe().pH
```

```python
# Bin edges that will be used to "cut" the data into groups
bin_edges = [2.72, 3.11, 3.21, 3.32, 4.01] # Fill in this list with five values you just found
```

```python
# Labels for the four acidity level groups
bin_names = ['high', 'mod_high', 'medium', 'low'] # Name each acidity level category
```

```python
# create a new column called acidity levels using pandas cut function
df_wine['acidity levels'] = pd.cut(df_wine.pH ,bin_edges ,labels = bin_names)
```

```python
# checking the new dataframe
df_wine.head()
```

```python
# Find the mean quality of each acidity level with groupby
df_wine.groupby('acidity levels').mean().quality
```

# Drawing Conclusions Using Query
Q1: Do wines with higher alcoholic content receive better ratings? **yes**
(To answer this question, use query to create two groups of wine samples:)

Low alcohol (samples with an alcohol content less than the median)
High alcohol (samples with an alcohol content greater than or equal to the median)
Then, find the mean quality rating of each group.

Q2: Do sweeter wines (more residual sugar) receive better ratings?
Similarly, use the median to split the samples into two groups by residual sugar and find the mean quality rating of each group.

```python
# loading the csv file into the dataframe
df_wine = pd.read_csv('winequality_edited.csv')
```

```python
# get the median amount of alcohol content
df_wine.alcohol.median()
```

```python
# select samples with alcohol content less than the median
low_alcohol = df_wine.query("alcohol < 10.3")

# select samples with alcohol content greater than or equal to the median
high_alcohol =df_wine.query("alcohol >= 10.3")

# ensure these queries included each sample exactly once
num_samples = df_wine.shape[0]
num_samples == low_alcohol['quality'].count() + high_alcohol['quality'].count() # should be True
```

```python
 # get mean quality rating for the low alcohol and high alcohol groups
print(low_alcohol.mean().quality)
print(high_alcohol.mean().quality) # it seems that wines with higher quality rating receive better ratings
```

### Do sweeter wines receive better ratings?

```python
# get the median amount of residual sugar
df_wine['residual sugar'].median()
```

```python
# select samples with residual sugar less than the median
low_sugar = df_wine[df_wine['residual sugar'] < 3]

# select samples with residual sugar greater than or equal to the median
high_sugar = df_wine[df_wine['residual sugar'] >= 3]


# ensure these queries included each sample exactly once
num_samples == low_sugar['quality'].count() + high_sugar['quality'].count() # should be True
```

```python
# get mean quality rating for the low sugar and high sugar groups
low_sugar.quality.mean() ,high_sugar.quality.mean() # seems like wines with high sugar receive better ratings
```

# Plotting with Matplotlib¶
Use Matplotlib to create bar charts that visualize the conclusions you made with groupby and query.

## 1: Do wines with higher alcoholic content receive better ratings?¶
Create a bar chart with one bar for low alcohol and one bar for high alcohol wine samples. 

```python
# get the median amount of alcohol content
df_wine.alcohol.median()
```

```python
# select samples with alcohol content less than the median
low_alcohol = df_wine.query("alcohol < 10.3")

# select samples with alcohol content greater than or equal to the median
high_alcohol =df_wine.query("alcohol >= 10.3")

# ensure these queries included each sample exactly once
num_samples = df_wine.shape[0]
num_samples == low_alcohol['quality'].count() + high_alcohol['quality'].count() # should be True

# mean quality of low and high rating samples

mean_quality_low = low_alcohol['quality'].mean()
mean_quality_high = high_alcohol['quality'].mean()
```

```python
# Create a bar chart with proper labels
plt.figure(figsize = (10 ,10))
height = [mean_quality_low ,mean_quality_high]
bars = ['Low', 'High']
y_pos = np.arange(len(bars))
 
# Create bars
plt.bar(y_pos, height)
 
# Rotation of the bars names
plt.xticks(y_pos, bars, rotation=90)
 
# Custom the subplot layout
plt.subplots_adjust(bottom=0.4, top=0.99)
plt.title('Average Quality Ratings by Alcohol Content')
plt.xlabel('Alcohol Content')
plt.ylabel('Average Quality Rating');
 
# Show graphic
plt.show()
```

## Plotting Wine Type and Quality with Matplotlib

Create arrays for red bar heights white bar heights
Remember, there's a bar for each combination of color and quality rating. Each bar's height is based on the proportion of samples of that color with that quality rating.

Red bar proportions = counts for each quality rating / total # of red samples

White bar proportions = counts for each quality rating / total # of white samples

```python
# get counts for each rating and color
color_counts = df_wine.groupby(['color', 'quality']).count()['pH']
color_counts
```

```python
# get total counts for each color
color_totals = df_wine.groupby('color').count()['pH']
color_totals
```

```python
# get proportions by dividing red rating counts by total # of red samples
red_proportions = color_counts['red'] / color_totals['red']
red_proportions
```

```python
# get proportions by dividing white rating counts by total # of white samples
white_proportions = color_counts['white'] / color_totals['white']
white_proportions
```

## Plot proportions on a bar chart
Set the x coordinate location for each rating group and and width of each

```python
ind = np.arange(len(red_proportions))  # the x locations for the groups
width = 0.35       # the width of the bars
```

Now let’s create the plot.

```python
# plot bars
red_bars = plt.bar(ind, red_proportions, width, color='r', alpha=.7, label='Red Wine')
white_bars = plt.bar(ind + width, white_proportions, width, color='w', alpha=.7, label='White Wine')

# title and labels
plt.ylabel('Proportion')
plt.xlabel('Quality')
plt.title('Proportion by Wine Color and Quality')
locations = ind + width / 2  # xtick locations
labels = ['3', '4', '5', '6', '7', '8', '9']  # xtick labels
plt.xticks(locations, labels)
```
